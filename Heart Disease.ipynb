```python
import pandas as pd
```

```python
# load csv 
heart_disease_data = pd.read_csv("./heart.csv")
heart_disease_data.head()
```

```python
training_data = heart_disease_data.drop(["target"], axis=1)
training_data.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(training_data, heart_disease_data.target, test_size=0.3, random_state=109)

X_train.head()
```

```python
from sklearn import svm

#create svm classifier
classifier = svm.SVC(kernel='linear')

#train classifier with training data
classifier.fit(X_train, y_train)

#predict
y_pred = classifier.predict(X_test)

y_pred
```
